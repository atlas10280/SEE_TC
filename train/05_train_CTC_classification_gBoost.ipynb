```python
##### Dependencies
import sys
import numpy as np

import sys
sys.path.append('../src/') 
import SEE_TC as ctc
```

```python
# Load cell embedings and review data structure in UMAP
path_data_encoder = "/mnt/local/data2/Bootsma/2D_CTC/src/analysis/publication_code/train_data/"

embeddings_ID = np.load(path_data_encoder+"03_train_CTC_classification_gBoost.CTC_latent.npy")
embeddings_OD = np.load(path_data_encoder+"03_train_CTC_classification_gBoost.exLow_latent.npy")
embeddings_Ex = np.load(path_data_encoder+"03_train_CTC_classification_gBoost.exHigh_latent.npy")

embeddings_plot = np.concatenate([embeddings_ID,embeddings_OD, embeddings_Ex], axis = 0)
labels_plot = np.concatenate([[0]*embeddings_ID.shape[0],[1]*embeddings_OD.shape[0],[2]*embeddings_Ex.shape[0]])
ctc.visualize_embeddings(embeddings_plot, labels_plot, method='umap')
```

```python
##### Train GB classifier here
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report


X = np.concatenate([embeddings_ID,embeddings_OD, embeddings_Ex], axis = 0)
y = np.concatenate([[0]*embeddings_ID.shape[0],[1]*embeddings_OD.shape[0],[1]*embeddings_Ex.shape[0]])

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
print(y_train.shape, y_test.shape)
ctc.count_labels(y_train)
ctc.count_labels(y_test)

# Train the model
clf = GradientBoostingClassifier()
clf.fit(X_train, y_train)

# Predict and evaluate
y_pred = clf.predict(X_test)
y_pred_prob = clf.predict_proba(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print(classification_report(y_test, y_pred))
```
